```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-4-mini-instruct",
                                             device_map="cuda",
                                             torch_dtype="auto",
                                             trust_remote_code=True)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-4-mini-instruct")
```

```python
prompt = "Escreva um email pedindo desculas a Sarah sobre trágico acidente ocorrido. Explique como ocorrey.<|assistant|>"
```

```python
tokenized_input = tokenizer(prompt, return_tensors="pt").to("cuda")
```

```python
tokenized_input
```

```python
generation_output = model.generate(input_ids=tokenized_input['input_ids'],
                                   attention_mask=tokenized_input['attention_mask'],
                                   max_new_tokens=20)
```
